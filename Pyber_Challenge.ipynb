# Pyber Challenge

## Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

## Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
driver_count
```

```python
#  3. Get the total amount of fares for each city type
total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare
```

```python
#  4. Get the average fare per ride for each city type. 
average_ridefare = total_fare/total_rides
average_ridefare
```

```python
# 5. Get the average fare per driver for each city type. 
average_driverfare = total_fare/driver_count
average_driverfare
```

```python
#  6. Create a PyBer summary DataFrame. 
PyBer_summary_df = pd.DataFrame({
          "Total Rides" : total_rides,
          "Total Drivers": driver_count,
          "Total Fares": total_fare,
          "Average Fare per Ride": average_ridefare,
          "Average Fare per Driver": average_driverfare})
PyBer_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
PyBer_summary_df.index.name = None
```

```python
#  8. Format the columns.
PyBer_summary_df["Total Fares"] = PyBer_summary_df["Total Fares"].map("${:,.2f}".format)
PyBer_summary_df["Average Fare per Ride"]= PyBer_summary_df["Average Fare per Ride"].map("${:.2f}".format)
PyBer_summary_df["Average Fare per Driver"]= PyBer_summary_df["Average Fare per Driver"].map("${:.2f}".format)
PyBer_summary_df
```

```python
# Import the style from Matplotlib.
from matplotlib import style
# Create a font library
titlefont = {"fontname":"Times New Roman"}
labelfont = {"fontname":"Arial"}
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
PyBer_summary_df.plot(kind = "bar", figsize=(10,6))
plt.title("Total Rides and drivers in Different City Types ",fontsize = 18,color="black",**titlefont)
plt.xlabel("City Types",fontsize = 15,color="black",**labelfont)
plt.xticks(rotation = 45)
plt.ylabel("Number of Rides and Drivers",fontsize = 15,color="black",**labelfont)
plt.tight_layout()
plt.savefig("analysis/PyBer_summary.png")
plt.show()
```

### The number of drivers in the Urban city type is way higher (2405) than the suburban (490) and rural city types (78).
### Since the number of drivers are less in the rural area the average fare per driver is way higher in the rural are.

```python
## Deliverable 2. Create a multiple line plot that shows the total weekly of the fares for each type of city.
```

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_by_date = pyber_data_df.groupby(["type","date"]).sum()["fare"]
fare_by_date
fare_by_date_df = pd.DataFrame(fare_by_date)
fare_by_date_df.head(10)
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fare_by_date_df = fare_by_date_df.reset_index()
fare_by_date_df.head(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

fare_pivot_df = fare_by_date_df.pivot(index = "date",columns = "type", values = "fare")
fare_pivot_df.head(10)
```

### Seasonal Ride Fare between January to April end in 2019 using pivot.

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_pivot_loc_df = fare_pivot_df.loc["2019-01-01":"2019-04-29"]
fare_pivot_loc_df.head(10)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_pivot_loc_df.index = pd.to_datetime(fare_pivot_loc_df.index)
fare_pivot_loc_df.head(10)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_pivot_loc_df.info()
```

### Weekly Fare analysis between January to April in 2019.

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_weekly_df = fare_pivot_loc_df.resample("W").sum()
fare_weekly_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
from matplotlib import style
# Create a font library
titlefont = {"fontname":"Times New Roman"}
labelfont = {"fontname":"Arial"}
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fare_weekly_df.plot(figsize=(20,6),linewidth =2, marker = "*")
plt.title("Total Fare by City Type ",fontsize = 18,color="black",**titlefont)
plt.xlabel("Months",fontsize = 15,color="black",**labelfont)
plt.ylabel("Fare ($USD)",fontsize = 15,color="black",**labelfont)
# Add the legend.
# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
        loc= 1, title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)
# Everything stays inside layout
plt.tight_layout()
plt.savefig("analysis/PyBer_fare_summary1.png")
plt.show() 
```

### Object oriented interface 

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
from matplotlib import style
# Create a font library
titlefont = {"fontname":"Times New Roman"}
labelfont = {"fontname":"Arial"}
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
ax = fare_weekly_df.plot(figsize=(20,6),linewidth =2, marker = "*")
ax.set_title("Total Fare by City Type ",fontsize = 18,color="black",**titlefont)
ax.set_xlabel("Months",fontsize = 15,color="black",**labelfont)
ax.set_ylabel("Fare ($USD)",fontsize = 15,color="black",**labelfont)
# Add the legend.
# Create a legend
lgnd = ax.legend(fontsize="12", mode="Expanded",
        loc= 1, title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)
# Everything stays inside layout
plt.tight_layout()
plt.savefig("analysis/PyBer_fare_summary2.png")
plt.show()  
```
